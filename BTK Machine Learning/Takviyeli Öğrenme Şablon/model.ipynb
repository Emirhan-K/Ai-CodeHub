```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import random
```

```python
veri=pd.read_csv('Ads_CTR_Optimisation.csv')
#Bu veri seti bir reklam ödül sistemidir tıklanılan reklamlardan 1 veya 0 ödül alınır
```

```python
#Random modülü ile kaç puan alabileceğimize bakacağız

N = 10000
d=10
toplam=0
secilen=[]
for n in range(0,N):
    ad=random.randrange(d)#random bir şekilde reklam seçiyor 1-10 arası bir reklama tıklıyor gibi düşünebiliriz
    secilen.append(ad)#seçilen reklamlar listesine ekliyoruz
    odul=veri.values[n,ad] # ödül miktarını da n'inci satırın ad'ıncı sütunundaki değere bakarak belirliyoruz
    toplam=toplam+odul
    
print('Alınan Puan:',toplam)
plt.hist(secilen)
plt.show()
```

```python
#UCB metodu 
#bakalım UCB metodu kullanırsak random modülünden daha iyi bir sonuç elde edebilir miyiz
#Bu metod ödül-ceza sistemi ile işler eğer tıklanan reklam 1 ise o reklamın ağırlığı artar.

N=10000 #10.000 tıklama
d=10 #reklam
#Ri(n)
oduller=[0]*d #ilk başta butun reklamların ödülü sıfır
#Ni(n)
tiklamalar=[0]*d #o ana kadarki tıklamalar

toplam=0 #toplam ödül
secilen=[]
for n in range(1,N):
    ad=0 #secilen ilan
    max_ucb=0
    for i in range(0,d):
        if(tiklamalar[i]>0): #ilk seçimden sonrası için geçerli max ucb belirlenmesi
            ortalama=oduller[i]/tiklamalar[i]
            delta=math.sqrt(3/2*math.log(n)/tiklamalar[i])
            ucb=delta+ortalama
            
        else: #ilk tıklamadaki ucb değeri belirlenir ve bayağı büyük bir sayı olmalıdır
            ucb=N*10
        if(max_ucb<ucb):#max_ucb den büyük ucb çıkarsa
            max_ucb=ucb
            ad=i
            
    secilen.append(ad)
    tiklamalar[ad]=tiklamalar[ad]+1
    odul=veri.values[n,ad] #verilerin n. satırı 1 ise ödül 1
    oduller[ad]=oduller[ad]+odul
    toplam=toplam+odul
    
print('Alınan Puan:',toplam)
plt.hist(secilen)
plt.show()    
```

```python
#Thompson Sampling metodu
N=10000 #10.000 tıklama
d=10 #reklam
#Ni(n)
toplam=0 #toplam ödül
secilen=[]
birler=[0]*d
sifirlar=[0]*d
for n in range(1,N):
    ad=0 #secilen ilan
    max_th=0
    for i in range(0,d):
        rasbeta=random.betavariate(birler[i]+1,sifirlar[i]+1)
        if (rasbeta>max_th):
            max_th=rasbeta
            ad=i
            
            
    secilen.append(ad)
    odul=veri.values[n,ad] #verilerin n. satırı 1 ise ödül 1
    if(odul==1):
        birler[ad]=birler[ad]+1
    else:
        sifirlar[ad]=sifirlar[ad]+1
    
    toplam=toplam+odul
    
print('Alınan Puan:',toplam)
plt.hist(secilen)
plt.show()  
```
