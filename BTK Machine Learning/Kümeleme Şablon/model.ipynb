```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Clustering yani Kümeleme modele sınıfları veya sonuçları vermeden kendi kendine veriyi bölmesini istediğimiz bir makine 
#öğrenmesi alanıdır.Genellikle müşteri segmentasyonu, sağlık, görüntü işleme ve pazarlamada kullanılır.
#Örneğin tavsiye algoritmaları, reklam önerileri, dolandırıcılık tespiti ve ayrıca eksik verilerin tamamlanmasında da kullanılır
```

```python
veri=pd.read_csv('veriler.csv')
veri=veri[['Hacim','Maas']].values
x=veri
#sadece hacim ve maas kısmını aldık 2 boyutlu uzayda görselleştirmek kolay olsun diye. Ancak gerekli olan tüm sütunları almalıyız
```

```python
#K-Means 

#Bu algoritma veriyi kaç kümeye böleceğimiz bilgisini ister(K).Daha sonra rasgele K tane nokta seçer. Ardından bu noktalara 
#en yakın olan noktaları o kümeye dahil eder.Sonra ayrılan K tane kümenin merkezleri belirlenir ve önceki adımlar tekrar uygulanır
#Bu işlem stabil olana kadar devam eder. Amaç merkezleri arası uzaklık max ve veriler arası uzaklığın min olduğu kümeleri bulmaktır.
#Başlangıç noktası rassal olarak seçildiği için küme merkezlerinin nerede olacağı belli olmuyor ve istenmeyen sonuçlar oluşabiliyor.
#K sayısını seçerken WCSS değeri denilen bir değer var. K değeri 1 den başlayarak bir döngüye girer. Bu döngünün içinde algoritma
#uygulanır ve merkezler belirlenir. Her verinin merkezine olan uzaklıklarının kareleri alınır ve toplanır.
#WCSS değerleri belirlendikten sonra bir grafiğe dökülür ve dirsek noktası belirlenir. Tabi bu işlem K sayısı çok hassas ve belirlenemiyorsa kullanılır.

from sklearn.cluster import KMeans
km=KMeans(n_clusters=3, n_init=10)
model_km=km.fit(x)

#model eğitimi bu kadar şimdi WCSS değerlerine bakalım
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i, n_init=10, random_state=123)
    km.fit(x)
    wcss.append(km.inertia_)#bu da wcss değerlerini ölçen fonksiyon
    
#artık elimizde wcss değerleri var ve biz uygun k sayısını bulabiliriz
plt.plot(range(1,11),wcss)
plt.show()
#grafiğe baktığımızda 2 veya 4 uygun k değerleridir
#4 değerini deneyip nasıl sonuç elde edeceğiz görelim

km=KMeans(n_clusters=4, n_init=10, random_state=123)
pred=km.fit_predict(x)

plt.scatter(x[pred==0,0],x[pred==0,1],s=50,c='red')
plt.scatter(x[pred==1,0],x[pred==1,1],s=50,c='blue')
plt.scatter(x[pred==2,0],x[pred==2,1],s=50,c='green')
plt.scatter(x[pred==3,0],x[pred==3,1],s=50,c='black')
plt.show()
```

```python
#Hiyerarşik bölütleme (HC)
#Agglomerative ve Divisive olarak 2 yöntemi vardır.
#Agglomerative ilk başta her bir veriyi bir küme olarak alır. Daha sonra verileri birbirine en yakın olanları iki elemandan oluşan 
#kümeler halinde böler. Bu işlem işlemler sonunda tek bir küme kalana kadar devam eder.
#Divisive ise bunun tam tersidir.İlk başta tüm verilerin dahil olduğu bir veri kümesi olarak kabul edilir. Daha sonra bu verileri 
#ikiye bölünür.Bölüne bölüne her bir veri bir küme olana kadar işleme devam edilir.
#Yani Agglomerative parçadan bütüne, Divisive ise bütünden parçaya şeklinde kümeler.

from sklearn.cluster import AgglomerativeClustering
ac=AgglomerativeClustering(n_clusters=4, metric='euclidean',linkage='ward')
pred=ac.fit_predict(x)

plt.scatter(x[pred==0,0],x[pred==0,1],s=50,c='red')
plt.scatter(x[pred==1,0],x[pred==1,1],s=50,c='blue')
plt.scatter(x[pred==2,0],x[pred==2,1],s=50,c='green')
plt.scatter(x[pred==3,0],x[pred==3,1],s=50,c='black')
plt.show()
```

```python
#Dendrogram
#Hiyerarşik kümelemede bize k sayısı seçiminde yardım eder.
#Kümeler arası mesafe en yüksek sayıda olan k sayısını seçmek doğru olabilir.

import scipy.cluster.hierarchy as sch

dendrogram = sch.dendrogram(sch.linkage(x,method='ward'))
plt.show()

#aşağıdaki dendrogram yardımıyla k'yı 2 veya 4 seçmemizib daha doğru olabileceğini görebiliyoruz
```
