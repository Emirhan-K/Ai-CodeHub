```python
y = dataFrame["Type"].values
x = dataFrame.drop("Type",axis = 1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(units=30,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")
```

```python
model.fit(x=x_train, y=y_train, epochs=100,validation_data=(x_test,y_test),verbose=1)#epochs kaç kez verinin üzerinden geçileceği
```

```python
model.history.history
```

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
model = Sequential()

model.add(Dense(units=30,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")
```

```python
#EarlyStopping 
'''Early stopping, makine öğrenimi ve derin öğrenme modellerinin eğitim sürecinde aşırı uyum (overfitting) riskini azaltmak 
ve modelin genelleme performansını artırmak için kullanılan bir yöntemdir. Early stopping, modelin eğitimini, doğrulama veri 
setindeki performansı takip ederek durdurma kriterine dayanarak erken sonlandırır.

Genellikle, eğitim süreci boyunca modelin doğrulama veri setindeki performansı izlenir ve her epoch sonunda doğrulama hatası,
doğruluk veya başka bir performans metriği kullanılarak değerlendirilir. Eğer modelin doğrulama performansı belirli bir süre 
boyunca (örneğin, ardışık birkaç epoch boyunca) daha iyi hale gelmezse, eğitim süreci durdurulur. Bu, modelin aşırı uyumdan kaçınmasına ve daha iyi bir genelleme yeteneğine sahip olmasına yardımcı olur.'''
earlyStopping = EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(x=x_train, y=y_train, epochs = 700, validation_data = (x_test,y_test), verbose = 1, callbacks=[earlyStopping])
```

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
model = Sequential()

model.add(Dense(units=30,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")
```

```python
model.fit(x=x_train, y=y_train, epochs = 700, validation_data = (x_test,y_test), verbose = 1, callbacks=[earlyStopping])
```

```python
kayipDf = pd.DataFrame(model.history.history)
```

```python
kayipDf.plot()
```

```python
tahminlerimiz = model.predict(x_test)
tahminlerimiz = np.argmax(tahminlerimiz, axis=1)
```

```python
tahminlerimiz
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,tahminlerimiz))
```

```python
print(classification_report(y_true=y_test,y_pred=tahminlerimiz))
```
