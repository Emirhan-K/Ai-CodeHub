```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_excel('veri.xlsx')
```

```python
df.isnull().sum()
```

```python
plt.scatter(df["Sales"], df["TV"], s=df["TV"])
x = df["Sales"]
y = df["TV"]
slope, intercept = np.polyfit(x, y, 1)
plt.plot(x, slope * x + intercept, color='red')

plt.show()
```

```python
plt.scatter(df["Sales"], df["Radio"], s=df["Radio"])
x = df["Sales"]
y = df["Radio"]
slope, intercept = np.polyfit(x, y, 1)
plt.plot(x, slope * x + intercept, color='red')

plt.show()
```

```python
df.corr()['Sales'].sort_values()
```

```python
x=df.drop('Sales',axis=1).values
y=df['Sales'].values
```

```python
xtrain, xtest, ytrain, ytest= train_test_split(x,y,test_size=0.2,random_state=18)
```

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
model.score(xtest,ytest)
```

```python
tahmin = np.array([[230.1, 37.8, 69.2]])
model.predict(tahmin)
```
